## Leetcode 181. Employees Earning More Than Their Managers

```python
data = [(1, 'Joe', 70000, 3), (2, 'Henry', 80000, 4),
        (3, 'Sam', 60000, None), (4, 'Max', 90000, None)]
Employee = spark.createDataFrame(data, ['id','name','salary','managerId'])
Employee.createOrReplaceTempView('Employee')
```

| id | name  | salary | managerId |
|----|-------|--------|-----------|
| 1  | Joe   | 70000  | 3         |
| 2  | Henry | 80000  | 4         |
| 3  | Sam   | 60000  | NULL      |
| 4  | Max   | 90000  | NULL      |

> **Note:** Rows with `managerId = NULL` are managers themselves. The self-join matches `e.managerId → m.id` to compare salaries.

### Spark SQL

```python
spark.sql('''
  SELECT e.name AS Employee
  FROM Employee e JOIN Employee m ON e.managerId = m.id
  WHERE e.salary > m.salary
''').show()

```

### Spark DataFrame API

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
emp = Employee.alias('e')
mgr = Employee.alias('m')
result = emp.join(mgr, emp.managerId == mgr.id)\
        .filter(col("e.salary") > col("m.salary"))\
        .select(emp.name.alias('Employee'))
result.show()
```
