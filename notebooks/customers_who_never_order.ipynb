## Leetcode 183. Customers Who Never Order

```python
cust  = [(1,'Joe'),(2,'Henry'),(3,'Sam'),(4,'Max')]
orders= [(1,3),(2,1)]
Customers = spark.createDataFrame(cust,   ['id','name'])
Orders    = spark.createDataFrame(orders, ['id','customerId'])
Customers.createOrReplaceTempView('Customers')
Orders.createOrReplaceTempView('Orders')
Customers.show()
Orders.show()
```

Customers
| id | name  |
|----|-------|
| 1  | Joe   |
| 2  | Henry |
| 3  | Sam   |
| 4  | Max   |

Orders
| id | customerId |
|----|------------|
| 1  | 3          |
| 2  | 1          |

### Spark SQL

```python
spark.sql('''
          select c.name as Customers from Customers c left join Orders o on c.id = o.customerId
where o.id is null
          ''').show()
```

### Spark DataFrame API

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
cust = Customers.alias('c')
orders = Orders.alias('o')

result  = cust.join(orders, cust.id == orders.customerId, "left")\
    .filter(col('o.id').isNull())\
        .select(col('c.name').alias('Customers')).show()
```
