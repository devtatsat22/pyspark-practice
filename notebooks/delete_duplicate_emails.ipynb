## Leetcode 196. Delete Duplicate Emails

```python
data = [(1,'john@example.com'),(2,'bob@example.com'),(3,'john@example.com')]
Person = spark.createDataFrame(data, ['id','email'])
Person.createOrReplaceTempView('Person')
Person.show()
```

### Spark SQL

```python
# Recreate the view with only non-duplicate rows (keeping minimum id for each email)
spark.sql('''
select min(id), email from Person group by email
''').show()

```

### Spark DataFrame API

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
Person.groupBy("email").agg(min(col('id')).alias('id')).select('id', 'email').show()
```
